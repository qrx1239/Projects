```python
text = "   The mountain is cold tonight!   "
cleaned_text = text.strip().lower().replace("!", ".")
print(cleaned_text)  # Output: "the mountain is cold tonight."  
```

```python
cleaned_text = cleaned_text.replace("mountain", "shop")
```

```python
print(cleaned_text)  # Output: "the shop is cold tonight."
```

```python
capitalized_text = cleaned_text.title()
```

```python
print(capitalized_text)  # Output: "The Shop Is Cold Tonight."
```

```python
cleaned_text = cleaned_text.replace("the", "The")
```

```python
print(cleaned_text)  # Output: "the shop is cold tonight."
```
